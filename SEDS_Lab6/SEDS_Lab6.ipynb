````python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import matplotlib.pyplot as plt
import re
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import nltk

# Ensure stopwords are downloaded
nltk.download('stopwords')

# Challenge 3: Extract text from `a` tags on Wikipedia page
url = "https://en.wikipedia.org/wiki/Data_science"

def scrape_anchor_text(url):
    try:
        page = urlopen(url).read().decode("utf-8")
        soup = bs(page, "lxml")
        links = soup.find_all("a")
        return [link.text.strip() for link in links if link.text.strip()]
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return []

def basic_text_analysis(text_list):
    # Join text list and clean
    text = " ".join(text_list).lower()
    text = re.sub(r"[^a-zA-Z\s]", "", text)

    # Tokenize and remove stopwords
    words = text.split()
    stop_words = set(stopwords.words('english'))
    filtered_words = [word for word in words if word not in stop_words]

    # Count words using FreqDist
    fdist = FreqDist(filtered_words)

    # Get most common words
    common_words = fdist.most_common(20)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.bar([word[0] for word in common_words], [word[1] for word in common_words])
    plt.title("Most Common Words in Anchor Tags (Stopwords Removed)")
    plt.xlabel("Words")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()

# Scrape and analyze
anchor_text = scrape_anchor_text(url)
basic_text_analysis(anchor_text)
````

````python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://www.goodreads.com/quotes/tag/motivational?page="
pages = [1, 2, 3, 4, 5]

header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win)'
}

quotes = []
authors = []
books = []
tags = []
likes = []

for page in pages:
    url = base_url + str(page)
    response = requests.get(url, headers=header)
    soup = BeautifulSoup(response.text, 'html.parser')

    quote_blocks = soup.find_all("div", {"class": "quote mediumText"})

    for quote_block in quote_blocks:
        quote_text = quote_block.find("div", {"class": "quoteText"})
        if quote_text:
            quote = quote_text.text.strip().split("\n")[0].strip()
            quotes.append(quote)

        author_or_title = quote_block.find("span", {"class": "authorOrTitle"})
        if author_or_title:
            author = author_or_title.text.strip()
            authors.append(author)
        
        book_title = quote_block.find("a", {"class": "authorOrTitle"})
        if book_title:
            books.append(book_title.text.strip())
        else:
            books.append("N/A")  

        tag_section = quote_block.find("div", {"class": "greyText smallText left"})
        if tag_section:
            tag_list = [tag.text for tag in tag_section.find_all("a")]
            tags.append(", ".join(tag_list))
        else:
            tags.append("N/A")

        like_section = quote_block.find("div", {"class": "quoteFooter"}).find("a", {"class": "smallText"})
        if like_section:
            like_count = int(like_section.text.strip().split()[0])
            likes.append(like_count)
        else:
            likes.append(0)


# print(quotes)

df = pd.DataFrame({
    "Quote": quotes,
    "Author/Book Title": authors,
    "Book": books,
    "Tags": tags,
    "Likes": likes
})

print(df.head())
````

````python
from urllib.request import urlopen
from bs4 import BeautifulSoup as bs
import matplotlib.pyplot as plt
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import nltk
import string

# Ensure stopwords are downloaded
nltk.download('stopwords')

# URLs for scraping Motivational quotes
urls = [
    "https://www.goodreads.com/quotes/tag/motivational?page=1",
    "https://www.goodreads.com/quotes/tag/motivational?page=2",
    "https://www.goodreads.com/quotes/tag/motivational?page=3",
    "https://www.goodreads.com/quotes/tag/motivational?page=4",
    "https://www.goodreads.com/quotes/tag/motivational?page=5"
]

# Scrape quotes, authors, book titles, tags, and likes
def scrape_motivational_quotes(url):
    try:
        page = urlopen(url).read().decode("utf-8")
        soup = bs(page, "lxml")

        quotes = []
        for quote_div in soup.find_all("div", {"class": "quote"}):
            # Extract quote text
            quote_text = quote_div.find("div", {"class": "quoteText"}).text.strip()
            quote_text = re.sub(r"\s+―.*", "", quote_text).strip()

            # Extract author or book title
            author_or_book = quote_div.find("span", {"class": "authorOrTitle"})
            author_or_book = author_or_book.text.strip() if author_or_book else "Unknown"
            
            book_title = quote_div.find("a", {"class": "authorOrTitle"})
            book_title=book_title.text.strip() if book_title else ""

            # Extract tags
            tags_div = quote_div.find("div", {"class": "greyText smallText left"})
            tags = tags_div.text.strip() if tags_div else ""

            # Extract likes
            likes_div = quote_div.find("a", {"class": "smallText"})
            likes = int(likes_div.text.split()[0]) if likes_div else 0

            quotes.append({
                "Quote": quote_text,
                "Author/Book": author_or_book,
                "Book_Titlte":book_title,
                "Tags": tags,
                "Likes": likes
            })

        return quotes
    except Exception as e:
        print(f"Error scraping {url}: {e}")
        return []

# Collect all data into a DataFrame
all_quotes = []
for url in urls:
    all_quotes.extend(scrape_motivational_quotes(url))

quotes_df = pd.DataFrame(all_quotes)

# Perform basic text analysis
stop_words = set(stopwords.words('english'))

def analyze_text(column_name):
    text = " ".join(quotes_df[column_name].dropna()).lower()
    translator = str.maketrans("", "", string.punctuation + string.digits + "`[]“”–’")
    text = text.translate(translator).replace("\n", " ").strip()

    # Tokenize and remove stopwords
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words and word.strip()]

    # Count words using FreqDist
    fdist = FreqDist(filtered_words)

    # Get most common words
    common_words = fdist.most_common(20)

    # Plot
    plt.figure(figsize=(10, 6))
    plt.bar([word[0] for word in common_words], [word[1] for word in common_words])
    plt.title(f"Most Common Words in {column_name}")
    plt.xlabel("Words")
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)
    plt.show()

# Analyze quotes text
analyze_text("Quote")

# Visualize number of likes by author/book
# author_likes = df.groupby("Author/Book Title")["Likes"].sum().reset_index()
# top_authors = author_likes.nlargest(10, "Likes")


author_likes = quotes_df.groupby("Author/Book")["Likes"].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
author_likes.head(10).plot(kind="bar")
plt.title("Top Authors/Books by Likes")
plt.xlabel("Author/Book")
plt.ylabel("Total Likes")
plt.xticks(rotation=45)
plt.show()

# Save DataFrame to CSV
quotes_df.to_csv("motivational_quotes.csv", index=False)
````
