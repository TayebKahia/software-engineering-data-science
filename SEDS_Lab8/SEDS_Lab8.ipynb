#  Bayes' law
Use Bayes’ law to calculate the probability of getting a data science job if you’ve gotten an interview for the job. This could be written P(get the DS job | interview). You’ll have to use Bayesian probability methods (your intuition or beliefs) to assign values to the different components of Bayes’ law.

`Hint:`
* Usually the probability of getting a job from applying is very low, around 1%. This is based on experience and reading about how most people apply to 100 jobs in order to get 1 job. 
* P(interview | get job) is 99%. Almost no one will hire someone they didn't interview. 
* P(interview | !get job) will be set to 10%. We might interview 10% of candidates, and only one gets the job.

***Bayes' Theorem states:***
$$ 
P(A|B)=\frac{P(B|A)  P(A)}{P(B)} $$
**SO:**
$$
P(\text{get the DS job } | \text{ interview}) = \frac{P(\text{interview }|\text{ get the DS job })P(\text{ get the DS job })}{P(\text{ interview })}
$$
we have:

$ P(\text{get the DS job }) = 0.01 \\ $
$ P(\text{interview }|\text{ get the DS job }) = 0.99 \\ $
$ P(\text{interview }| \text{ not get the DS job }) = 0.1 \\ $
$ P(\text{interview }) =  P(\text{get the DS job })P(\text{interview }|\text{ get the DS job }) + P(\text{get the DS job })P(\text{interview }| \text{ not get the DS job }) $



```python
#Your Solution
p_getDsJob = 0.01
p_inter__getDsJob = 0.99
p_inter = 0.1*0.99 + 0.1*0.1

p_getDsJob__inter = (p_inter__getDsJob*p_getDsJob)/(p_inter)
p_getDsJob__inter
```

```python
miso_data  = pd.read_csv('data/miso_wind_data.csv')
miso_data.sample(100)
```

```python
mwh_data = miso_data['MWh'].dropna().astype(float)
mwh_data = mwh_data[mwh_data >= 0]
mwh_data.head()
```

```python
import scipy.stats as stats
result = {}
```

### Test Normal Distribution
**Formula:**
$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^{2}}}
$$
**Parameters:**
- $\mu$ (mean): Determines the center of the distribution.
- $\sigma$ (standard deviation): Measures the spread or dispersion of the data.

```python
x = np.linspace(mwh_data.min(), mwh_data.max(), 100)
params_norm = stats.norm.fit(mwh_data)
pdf = stats.norm.pdf(x, *params_norm)
plt.plot(x, pdf, label=f'Normal PDF', color='red')
plt.hist(mwh_data, bins=30, density=True, alpha=0.6, color='blue', label='Data')
plt.legend()
plt.title('Fitted Distributions')
plt.xlabel('MWh')
plt.ylabel('Density')
plt.show()
```

```python
mu, std = params_norm
ks_stat_norm, p_value_norm = stats.kstest(mwh_data, 'norm', args=(mu, std))
print("Normal Distribution Parameters:")
print(f"Mean (μ): {mu}")
print(f"Standard Deviation (σ): {std}")
print(f"KS Statistic: {ks_stat_norm}")
print(f"P-value: {p_value_norm}")
result['norm'] = (ks_stat_norm, p_value_norm)
```

### Exponential Distribution
**Formula:**
$$ f(x;\lambda) = \lambda e^{-\lambda x}, \ \ x \ge  0 $$
**Parameters:**
- $\lambda$ (rate parameter): Inverse of the mean $(\lambda = 1 / \text{mean})$

```python
params_exp = stats.expon.fit(mwh_data)
pdf = stats.expon.pdf(x, *params_exp)
plt.plot(x, pdf, label=f'Exponential PDF', color='red')
plt.hist(mwh_data, bins=30, density=True, alpha=0.6, color='blue', label='Data')
plt.legend()
plt.title('Fitted Distributions')
plt.xlabel('MWh')
plt.ylabel('Density')
plt.show()
```

```python
loc_exp, scale_exp = params_exp
ks_stat_exp, p_value_exp = stats.kstest(mwh_data, 'expon', args=(loc_exp, scale_exp))
print("Exponential Distribution Parameters:")
print(f"Location (loc): {loc_exp}")
print(f"Scale (β): {scale_exp}")
print(f"KS Statistic: {ks_stat_exp}")
print(f"P-value: {p_value_exp}")
result['expon'] = (ks_stat_exp, p_value_exp)
```

### Test Weibull Distribution
**Formula:**
$$ f(x;k,\lambda) = \frac{k}{\lambda}\Big(\frac{x}{\lambda}\Big)^{k-1}e^{-\big(\frac{x}{\lambda}\big)^k}, \ \ x\ge 0 $$
**Test:**
- $k$ (shape parameter): Determines the shape of the distribution
  - $k<1$ Decreasing failure rate (e.g., early failures).
  - $k=1$: Constant failure rate (exponential distribution).
  - $k>1$: Increasing failure rate (e.g., aging or wear-out failures).
- $\lambda$ (scale parameter): Stretches or shrinks the distribution.

```python
params_weibull = stats.weibull_min.fit(mwh_data)

plt.figure(figsize=(10, 6))
plt.hist(mwh_data, bins=30, density=True, alpha=0.7, label='Actual Data')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
pdf = stats.weibull_min.pdf(x, *params_weibull)
plt.plot(x, pdf, 'k', linewidth=2, label='Weibull Fit')
plt.title('Weibull Distribution Fit')
plt.xlabel('MWh')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
c, loc_weibull, scale_weibull = params_weibull
ks_stat_weibull, p_value_weibull = stats.kstest(mwh_data, 'weibull_min', args=(c, loc_weibull, scale_weibull))
print("Weibull Distribution Parameters:")
print(f"Shape (c): {c}")
print(f"Location (loc): {loc_weibull}")
print(f"Scale (β): {scale_weibull}")
print(f"KS Statistic: {ks_stat_weibull}")
print(f"P-value: {p_value_weibull}")
result["weibull_min"] = (ks_stat_weibull, p_value_weibull)
```

### Test Gamma Distribution
**Formula:**
$$ f(x;k,\theta) = \frac{x^{k-1}e^{-x/\theta}}{\theta^k\Gamma(k)}, \ \ x\ge 0 $$
**Parameters:**
- $k$ (shape parameter): Determines the shape of the distribution.
- $θ$ (scale parameter): Scales the distribution horizontally.

```python
params_gamma = stats.gamma.fit(mwh_data)
pdf = stats.gamma.pdf(x, *params_gamma)
plt.plot(x, pdf, label=f'Gamma PDF', color='red')
plt.hist(mwh_data, bins=30, density=True, alpha=0.6, color='blue', label='Data')
plt.legend()
plt.title('Fitted Distributions')
plt.xlabel('MWh')
plt.ylabel('Density')
plt.show()
```

```python
alpha, loc_gamma, beta = params_gamma
ks_stat_gamma, p_value_gamma = stats.kstest(mwh_data, 'gamma', args=(alpha, loc_gamma, beta))
print("Gamma Distribution Parameters:")
print(f"Shape (α): {alpha}")
print(f"Location (loc): {loc_gamma}")
print(f"Scale (β): {beta}")
print(f"KS Statistic: {ks_stat_gamma}")
print(f"P-value: {p_value_gamma}")
result['gamma'] = (ks_stat_gamma, p_value_gamma)
```

| Distribution      | Shape                   | Domain            | Common Use Case                           |
|-------------------|-------------------------|-------------------|-------------------------------------------|
| Normal            | Bell-shaped, symmetric | ]-∞, ∞[           | Modeling symmetric data like heights.     |
| Exponential       | Decaying exponentially | [0, ∞[            | Waiting times (e.g., inter-arrival times).|
| Weibull           | Flexible (depends on k)| [0, ∞[            | Reliability or failure modeling.          |
| Gamma             | Skewed right           | [0, ∞[            | Waiting times with bursts or clusters.    |

```python
def find_best_distribution(result):
    # Sort distributions by KS statistic (lower is better)
    # And by p-value (higher is better)
    sorted_distributions = sorted(
        result.items(), 
        key=lambda x: (x[1][0], -x[1][1])
    )
    
    print("Distribution Ranking:")
    for rank, (name, (ks_stat, p_value)) in enumerate(sorted_distributions, 1):
        print(f"{rank}. {name}:")
        print(f"   KS Statistic: {ks_stat}")
        print(f"   P-value: {p_value}")
    
    best_distribution = sorted_distributions[0][0]
    print(f"\nBest Distribution: {best_distribution}")
    
    return best_distribution
```


# Bootstrap Sampling
Apply bootstrap sampling to the MISO wind power generation data and plot the distribution of the mean wind power generation values.

```python
n_bootstrap = 1000  
bootstrap_means = []
for _ in range(n_bootstrap):
    sample = np.random.choice(mwh_data, size=len(mwh_data), replace=True)
    bootstrap_means.append(np.mean(sample))
    
plt.figure(figsize=(10, 6))
plt.hist(bootstrap_means, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Bootstrap Means for Wind Power Generation')
plt.xlabel('Mean MWh')
plt.ylabel('Density')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Number of bootstrap samples
n_bootstrap_samples = 1000
sample_size = len(mwh_data)

# Perform bootstrap sampling
bootstrap_means = []
for _ in range(n_bootstrap_samples):
    bootstrap_sample = np.random.choice(mwh_data, size=sample_size, replace=True)
    bootstrap_means.append(np.mean(bootstrap_sample))

# Convert bootstrap means to a NumPy array
bootstrap_means = np.array(bootstrap_means)

# Plot the distribution of bootstrap means
plt.figure(figsize=(8, 5))
plt.hist(bootstrap_means, bins=50, density=True, alpha=0.7, color='blue')
plt.title("Bootstrap Sampling Distribution of Mean Wind Power Generation")
plt.xlabel("Mean MWh")
plt.ylabel("Density")
plt.axvline(np.mean(bootstrap_means), color='red', linestyle='dashed', linewidth=2, label="Mean of Bootstrap Means")
plt.legend()
plt.show()

# Summary statistics of bootstrap means
bootstrap_mean = np.mean(bootstrap_means)
bootstrap_std = np.std(bootstrap_means)

print(f"Bootstrap Mean: {bootstrap_mean:.2f}")
print(f"Bootstrap Standard Deviation: {bootstrap_std:.2f}")
```

```python
lower_ci = np.percentile(bootstrap_means, 2.5)
upper_ci = np.percentile(bootstrap_means, 97.5)

print(f"95% Confidence Interval for the Mean: ({lower_ci:.2f}, {upper_ci:.2f})")
```

```python
import pandas as pd
import numpy as np
from scipy.stats import norm, expon, weibull_min, gamma, kstest
import matplotlib.pyplot as plt

# # Load the data
# file_path = '/mnt/data/miso_wind_data.csv'
# miso_data = pd.read_csv(file_path)

# # Data cleaning
# miso_data['MWh'] = pd.to_numeric(miso_data['MWh'], errors='coerce')
# miso_data = miso_data.dropna(subset=['MWh'])
# mwh_data = miso_data['MWh']
# mwh_data = mwh_data[mwh_data >= 0]

# Fit and plot distributions
def fit_and_plot(data, dist, dist_name):
    params = dist.fit(data)
    x = np.linspace(data.min(), data.max(), 1000)
    pdf_fitted = dist.pdf(x, *params)
    
    plt.figure(figsize=(8, 5))
    plt.hist(data, bins=50, density=True, alpha=0.6, color='gray', label='Histogram')
    plt.plot(x, pdf_fitted, label=f'{dist_name} Fit', linewidth=2)
    plt.title(f'{dist_name} Distribution Fit')
    plt.xlabel('MWh')
    plt.ylabel('Density')
    plt.legend()
    plt.show()
    
    return params

# Fit and plot Normal distribution
normal_params = fit_and_plot(mwh_data, norm, "Normal")

# Fit and plot Exponential distribution
exponential_params = fit_and_plot(mwh_data, expon, "Exponential")

# Fit and plot Weibull distribution
weibull_params = fit_and_plot(mwh_data, weibull_min, "Weibull")

# Fit and plot Gamma distribution
gamma_params = fit_and_plot(mwh_data, gamma, "Gamma")

# Goodness-of-fit evaluation using Kolmogorov-Smirnov test
def ks_test(data, dist, params):
    cdf = lambda x: dist.cdf(x, *params)
    stat, p_value = kstest(data, cdf)
    return stat, p_value

# KS test for each distribution
ks_results = {
    "Normal": ks_test(mwh_data, norm, normal_params),
    "Exponential": ks_test(mwh_data, expon, exponential_params),
    "Weibull": ks_test(mwh_data, weibull_min, weibull_params),
    "Gamma": ks_test(mwh_data, gamma, gamma_params),
}

# Display KS test results
for dist_name, (stat, p_value) in ks_results.items():
    print(f"{dist_name} Distribution: KS Statistic = {stat:.4f}, p-value = {p_value:.4f}")

# Identify the best-fit distribution
best_fit = min(ks_results, key=lambda k: ks_results[k][0])
print(f"\nBest-fitting distribution: {best_fit}")
```
